```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
sns.lmplot(x='total_bill',y='tip',data=df,legend=True,hue='sex') #use Fit_reg=False to see normal scatterplot
plt.show()
```

```
sns.stripplot(x='day',y='total_bill',data=df)
plt.show()
```

```
sns.swarmplot(x='day',y='total_bill',data=df)
plt.show()
```

```
sns.countplot(x='sex',data=df)
plt.show()
```

```
sns.distplot(df['total_bill'])
plt.show()
```

```
df['total_bill'].skew() #skewness is +ve right tilt
```

```
df['total_bill'].kurtosis()
```

```
sns.boxplot(x='total_bill',data=df)
plt.show()
```

```
sns.boxplot(x='day',y='total_bill',data=df)
plt.show()
```

```
sns.jointplot(x='tip',y='total_bill',data=df)
plt.show()
```

```
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
plt.show()
```
